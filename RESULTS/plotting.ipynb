```julia
using Plots;
using JLD;
using Printf;
```

# Create Reinforcement Learning Environment

```julia
# number of steps
steps = 1_500_000;
```

```julia
num_model = Int(steps / 10_000);
num_test = 200; # TODO: change as desired 
```

```julia
fontsize = 12
```

```julia
data_path = pwd() * "/plots/"; # TODO: path to desired folder (where to load data from)
fig_path = pwd() * "/plots/combi/"; # TODO: path to desired folder (where to safe figures)
```

```julia
results_path = data_path * "slow/"

iterations = load(results_path * "iterations.jld")["data"];

avg_compl_time_slow = load(results_path * "avg_comp_time.jld")["data"];
avg_velocity_slow = load(results_path * "avg_velocity.jld")["data"];
rewards_slow = load(results_path * "reward.jld")["data"];
success_rate_slow = load(results_path * "success_rate.jld")["data"];


results_path = data_path * "fast/"

avg_compl_time_fast = load(results_path * "avg_comp_time.jld")["data"];
avg_velocity_fast = load(results_path * "avg_velocity.jld")["data"];
rewards_fast = load(results_path * "reward.jld")["data"];
success_rate_fast = load(results_path * "success_rate.jld")["data"];
```

```julia
ticks = collect(0:250_000:1_500_000);
ticklabels = [ @sprintf("%1.2fx10‚Å∂",iterations/1000000) for iterations in ticks ];
```

```julia
plot(iterations, rewards_slow, 
    xlabel="Iterations", 
    ylabel="Reward", 
    label = "slow mode", 
    legend = :right, 
    legendfontsize=fontsize,
    xtickfontsize=fontsize-4,
    ytickfontsize=fontsize,
    xguidefontsize=fontsize,
    yguidefontsize=fontsize,
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "reward_slow.svg")
```

```julia
plot(iterations, rewards_fast, 
    xlabel="Iterations", 
    ylabel="Reward", 
    label = "fast mode", 
    color = :orange, 
    legend = :right,   
    legendfontsize=fontsize,
    xtickfontsize=fontsize-4,
    ytickfontsize=fontsize,
    xguidefontsize=fontsize,
    yguidefontsize=fontsize,
    xformatter = :scientific,
)
plot!(xticks=(ticks,ticklabels))
```

```julia
savefig(fig_path * "reward_fast.svg")
```

```julia
plot(iterations, success_rate_slow, 
    xlabel="Iterations", 
    ylabel="Success Rate", 
    label = "slow mode", 
    legend = true, 
    legendfontsize=fontsize,
    xtickfontsize=fontsize-4,
    ytickfontsize=fontsize,
    xguidefontsize=fontsize,
    yguidefontsize=fontsize,
    xformatter = :scientific
)
plot!(iterations, success_rate_fast, 
    xlabel="Iterations", 
    ylabel="Success Rate", 
    label = "fast mode", 
    legend = :right, 
    color = :orange, 
    xformatter = :scientific
)
plot!(xticks=(ticks,ticklabels))
plot!(yticks=0:0.1:1)
```

```julia
savefig(fig_path * "success_rate.svg")
```
